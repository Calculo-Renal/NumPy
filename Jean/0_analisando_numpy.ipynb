# Analisando o NumPy

## O que é o NumPy?

O NumPy é uma biblioteca em Python que visa realizar operações matemáticas, simples ou complexas de forma rápida e/ou curta. A forma que a biblioteca realiza isso é através de `ndarray`s, estruturas imutáveis e de tipo homogêneo.

A forma como o NumPy acelera cálculos e simplifica código é criando uma interface reduzida e menos rigorosa para métodos, que são implementados da forma que for mais rápida (por exemplo, usando código pré-compilado em C).
